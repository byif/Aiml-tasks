# OBSERVATIONS:
- there are no missing values
- there are no duplicated values
- all columns are converted into numeric columns

```python

from sklearn.preprocessing import StandardScaler

# Function to clean percentage and currency values
def clean_numeric_column(col):
    return pd.to_numeric(col.replace({',': '', '%': '', '\$': ''}, regex=True), errors='coerce')
# Identify columns that need conversion
columns_to_clean = ["Business Tax Rate", "GDP", "Health Exp/Capita", "Tourism Inbound", "Tourism Outbound"]
data[columns_to_clean] = data[columns_to_clean].apply(clean_numeric_column)
# Select only numerical columns for scaling (excluding categorical ones like 'Country')
numerical_cols = data.select_dtypes(include=[np.number]).columns
# Apply StandardScaler
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data[numerical_cols]), columns=numerical_cols)
# Display the first few rows of the scaled dataset
print(scaled_data.head())
```

```python
# Build 3 clusters using KMeans Cluster algorithm
from sklearn.cluster import KMeans
clusters_new=KMeans(3,random_state=0)
clusters_new.fit(scaled_data)
```

```python
#import numpy as np
#import pandas as pd
#import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Compute Cluster Characteristics
cluster_summary = data.groupby("clusterid_new")[numerical_cols].mean()
print("Cluster Characteristics:\n", cluster_summary)

# Standardize data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[numerical_cols])

# Find optimal clusters using the Elbow Method
wcss = []
silhouette_scores = []
K_range = range(2, 15)  # Testing k from 2 to 15

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(scaled_data)
    
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_data, kmeans.labels_))

# Plot Elbow Curve with Log Scale
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")

plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, marker='s', linestyle='-', color='r')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score Analysis")

plt.show()
```
